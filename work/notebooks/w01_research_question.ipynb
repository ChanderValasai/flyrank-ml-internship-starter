## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

Decision improved: which existing pages a content reviewer should look at first, out of thousands, when refresh time is limited. Unit of analysis: one content item (content_id). Actor: a content/SEO reviewer. Action: they open the top-N pages in the ranked queue and decide, using the reason codes, whether to refresh, expand, protect, or monitor each one. Cost of a wrong call: if a page is wrongly ranked high, the reviewer spends limited time on it for no benefit — while a page that's actually declining may sit unreviewed. Neither error is catastrophic, but both cost reviewer time or missed traffic, which is why ranking quality (not just raw accuracy) is what matters here.
